```python
import pandas as pd
import jieba
from gensim.models.word2vec import Word2Vec
# 读入训练集文件
data = pd.read_csv('train.csv')
# 转字符串数组
corpus = data['comment'].values

# 分词，再重组为字符串数组
corpus = [jieba.lcut(corpus[index]
                          .replace("，", "")
                          .replace("!", "")
                          .replace("！", "")
                          .replace("。", "")
                          .replace("~", "")
                          .replace("；", "")
                          .replace("？", "")
                          .replace("?", "")
                          .replace("【", "")
                          .replace("】", "")
                          .replace("#", "")
                        ) for index in range(len(corpus))]
# 词向量模型训练
model = Word2Vec(corpus, sg=0, vector_size=300, window=5, min_count=3, workers=4)
#模型显示
print('模型参数：',model,'\n')
```

```python
    print('词向量：', model.wv['环境'])
    print('形状：', model.wv['环境'].shape)
```

```python
 print('与“好吃”最接近的3个词:',model.wv.most_similar('好吃', topn=3))
```

```python
print('相似度为=',model.wv.similarity('美味','好吃'),'\n')
print('相似度为=',model.wv.similarity('蟑螂','好吃'),'\n')
```

```python
print('最匹配的词是：',model.wv.most_similar(positive=['餐厅', '聚会'], negative=['安静'],topn=1),'\n')
```
